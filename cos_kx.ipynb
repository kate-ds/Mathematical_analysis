```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
k = 1
# y = np.cos(k*x)
plt.plot(x, np.cos(k*x), label='k=1')
k=3
plt.plot(x, np.cos(k*x), label='k=3')
plt.title('cos(kx)', size=20)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
