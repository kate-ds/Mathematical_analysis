### $$\frac{A}{x+2} + \frac{B}{x-1} = \frac{Ax - A + Bx + 2B}{(x+2)(x-1)} = \frac{x(A + B) + (2B -A)}{(x+2)(x-1)}$$

### $$\begin{cases}

### $$\int\limits_2^{+\infty} \frac{dx}{x^2+x-2} = -\frac{1}{3} \int\limits_2^{+\infty} \frac{dx}{x+2} + \frac{1}{3} \int\limits_2^{+\infty} \frac{dx}{x-1} = -\frac{1}{3} (\lim_{b\to +\infty} ln|b+2| - ln|4|) + \frac{1}{3} (\lim_{b\to +\infty} ln|b-1| - ln|1|) = \\ = \frac{1}{3} \Bigl( \lim_{b\to +\infty} ln|b-1| - \lim_{b\to +\infty} ln|b+2| + ln4 \Bigr) = \frac{1}{3} \Bigl( \lim_{b\to +\infty} ln\frac{|b-1|}{|b+2|} + ln4 \Bigr) = \frac{1}{3} (ln1 + ln4) = \frac{1}{3}ln4$$
