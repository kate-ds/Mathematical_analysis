#### $$1.\,\,\,\{a_n\}_{n=1}^\infty=2^n-n$$  

a. Исследовать их на монотонность:   
    $a_1 = 2^1 - 1 = 1$   
    $a_2 = 2^2 - 2 = 2$   
    $a_3 = 2^3 - 3 = 5$   
    $a_1 < a_2 < a_3$ - последовательность монотонно возрастает  

b. Исследовать на ограниченность:  
Оценка ограничения снизу при $n=1: \,\,\, a_1 = 2^1-1 = 1$  
Оценка ограничения снизу при $n\to \infty$:  
$$\lim_{n\to \infty} (2^n-n) = \lim_{n\to \infty} \Bigr((2^n-n)\cdot \frac{2^n+n}{2^n+n}\Bigl) = \lim_{n\to \infty} \Bigr( \frac{2^{2n}-n^2}{2^n+n} \Bigl) = \lim_{n\to \infty} \frac{2^n\Bigr(2^{2}-\frac{n^2}{2^n}\Bigl)}{2^n\Bigr(1-\frac{n}{2^n}\Bigl)} = \frac{4}{1} = 4$$  
Последовательность ограничена снизу в значении $1$, сверху в значении $4$

c. Найти пятый по счету член:  
$$a_5 = 2^5 - 5 = 27$$
------------------------------------------

------------------------------------------
#### $$2.\,\,\,\{b_n\}_{n=2}^\infty=\frac{1}{1 - n}$$ 

a. Исследовать их на монотонность:   
    $b_2 = \frac{1}{1 - 2} = -1$   
    $b_3 = \frac{1}{1 - 3} = -\frac{1}{2}$   
    $b_4 = \frac{1}{1 - 4} = -\frac{1}{3}$   
    $b_2 < b_3 < b_4$ - последовательность монотонно возрастает  

b. Исследовать на ограниченность:  
Оценка ограничения снизу при $n=2: \,\,\, b_2 = \frac{1}{1 - 2} = -1$  
Оценка ограничения снизу при $n\to \infty$:  
$$\lim_{n\to \infty} \frac{1}{1 - 2} = 0$$
Последовательность ограничена снизу в значении $-1$, сверху в значении $0$

c. Найти пятый по счету член:  
$$b_5 = \frac{1}{1 - 5} = -\frac{1}{4}$$

#### $$3.\,\,\,\{c_n\}_{n=1}^\infty=(-1)^n + \sqrt{2n}$$

a. Исследовать их на монотонность:   
    $c_1 = (-1)^{1} + \sqrt{2\cdot 1} = -1 + \sqrt{2} \approx 0,4$   
    $c_2 = (-1)^2 + \sqrt{2\cdot 2} = 1 + 2 = 3$   
    $c_3 = (-1)^3 + \sqrt{2\cdot 3} = -1 + \sqrt{6} \approx 1,45$   
    $c_1 < c_2 > c_3$ - последовательность немонотонная  

b. Исследовать на ограниченность:  
Оценка ограничения снизу при $n=1: \,\,\, c_1 = (-1)^{1} + \sqrt{2\cdot 1} = -1 + \sqrt{2}$  
Оценка ограничения снизу при $n\to \infty$:  
$$\lim_{n\to \infty} ((-1)^n + \sqrt{2n}) = \lim_{n\to \infty} \frac{((-1)^n + \sqrt{2n})\cdot ((-1)^n - \sqrt{2n})}{(-1)^n - \sqrt{2n}} = \frac{(-1)^{2n} - 2n}{(-1)^n - \sqrt{2n}} = \infty$$
Последовательность неограничена

c. Найти пятый по счету член:  
$$c_5 = (-1)^5 + \sqrt{2\cdot 5} = -1+ \sqrt{10} \approx 2,16$$

------------------------------------------
#### $$4.\,\,\,\{d_n\}_{n=1}^\infty=(-1)^{2n} + \frac{1}{n^2}$$  

a. Исследовать их на монотонность:   
    $d_1 = (-1)^{2} + \frac{1}{1^2} = 1 + 1 = 2$   
    $d_2 = (-1)^{4} + \frac{1}{2^2} = 1 + 0,25 = 1,25$   
    $d_3 = (-1)^{6} + \frac{1}{3^2} = 1 + \frac{1}{9} = 1\frac{1}{9}$   
    $d_1 > d_2 > d_3$ - последовательность монотонно убывает  

b. Исследовать на ограниченность:  
Оценка ограничения снизу при $n=1: \,\,\, d_1 = (-1)^{2} + \frac{1}{1^2} = 1 + 1 = 2$  
Оценка ограничения снизу при $n\to \infty$:  
$$\lim_{n\to \infty} \bigr((-1)^{2n} + \frac{1}{n^2}\bigl) = 1 + 0 = 1$$
Последовательность ограничена снизу в значении $2$, сверху в значении $1$

c. Найти пятый по счету член:  
$$d_5 = (-1)^{10} + \frac{1}{5^2} = 1 + \frac{1}{25} = 1,04$$

__3.__ *На языке Python предложить алгоритм вычисляющий численно предел с точностью $\varepsilon = 10^{-7}$
#### $$\lim_{n\to \infty} \frac{n}{\sqrt[n]{n!}}$$
__4.__ *Предложить оптимизацию алгоритма, полученного в задании 3, ускоряющую его сходимость.

```python
from scipy.special import factorial

eps = 0.0000001
lim=1

n = 2
a2 = 1

while lim > eps:
    a1 = a2
    a2 = n/(factorial(n)**(1/n))
    lim = abs(a1-a2)
    n += 1
print (f'Численное значение предела: {a2}\n',f'Погрешность: {lim}\n', f'Число шагов: {n-1}')
```

__4.__ Пользуясь критерием Коши, докажите сходимость последовательности:

### $$a_n=\frac{\sin1}{2}+\frac{\sin2}{2^2}+\frac{\sin3}{2^3}+..\frac{\sin n}{2^n}\,\,\,\,\,\Rightarrow$$

### $$\Bigl\{a_n\Bigr\}_{n=1}^\infty=\Bigl\{\frac{\sin1}{2}, \frac{\sin1}{2}+\frac{\sin2}{2^2}, ..., a_n,...\Bigr\}$$


$^*$ Какой член последовательности можно взять в качестве предела с точностью $\varepsilon=10^{-7}$?

Для доказательства сходимости по критерию Коши сходимости последовательности нужно найти такое $N(\varepsilon)$, чтобы выполнялось $|a_n - a_{n-k}| < \varepsilon$

$$|a_n - a_{n-k}| = \Bigl| \frac{\sin1}{2}+\frac{\sin2}{2^2}+\frac{\sin3}{2^3}+..\frac{\sin n}{2^n} - \Bigl( \frac{\sin1}{2}+\frac{\sin2}{2^2}+\frac{\sin3}{2^3}+..\frac{\sin (n+k)}{2^{n+k}}\Bigr)\Bigr| =
= \Bigl| \frac{\sin(n+1)}{2^{n+1}}+\frac{\sin(n+2)}{2^{n+2}}+..\frac{\sin(n+k)}{2^{n+k}}\Bigr| =\\ \frac{1}{2^n}\Bigl| \frac{1}{2}+\frac{1}{4}+..\frac{1}{2^k}\Bigr| \leqslant  \frac{1}{2^n} < \varepsilon 
\Rightarrow \frac{1}{2^{N(\varepsilon)}} = \varepsilon \Rightarrow N(\varepsilon) = \log_2{\frac{1}{\varepsilon}}$$

При $\varepsilon = 10^{-7}$ $N(\varepsilon) \approx 23,25. $  
$n > N(\varepsilon) \Rightarrow n = 24$
