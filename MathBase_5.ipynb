```python
# задаем две монеты, 1 - орел, 0 - решка  (подбрасываем 50 раз)
n = 5000
x = np.random.randint(0, 2, n)
y = np.random.randint(0, 2, n)

# вероятность того, что выпадут два орла
m = 0
for i in range(n):
    if (x[i] == 1) and (x[i] == y[i]):
        m += 1

p = m/n
p
```
