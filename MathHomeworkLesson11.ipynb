### ДЗ по теме “Ряды”

__1.__ Исследовать ряд на сходимость, используя признак д’Аламбера:

### $$\sum\limits_{n=1}^{+\infty}\frac{n^n}{(n!)^2}$$

### $$\lim_{n\to +\infty} \frac{a_{n+1}}{a_n} = \lim_{n\to +\infty} \frac{(n+1)^{n+1}}{((n+1)!)^2} \cdot \frac{(n!)^2}{n^n} = \lim_{n\to +\infty} \frac{(n+1)^n(n+1)}{(n!)^2(n+1)^2} \cdot \frac{(n!)^2}{n^n} = \lim_{n\to +\infty} \frac{1}{n+1} \cdot \Bigl(\frac{n + 1}{n}\Bigr)^n = 0$$

### $$q = 0 < 1 - ряд\,сходится $$

__2.__ Исследовать ряд на сходимость, используя радикальный признак Коши:

### $$\sum\limits_{n=1}^{+\infty}\frac{n}{2^n}$$

### $$\lim_{n\to +\infty} \sqrt[n] a_{n} =  \lim_{n\to +\infty} \sqrt[n] \frac{n}{2^n} = \lim_{n\to +\infty} \frac{\sqrt[n]n} {2} = \frac{1}{2}$$

### $$q = \frac{1}{2} < 1 - ряд\,сходится $$

__3.__ Исследовать ряд на сходимость, используя признак Лейбница:

### $$\sum\limits_{n=1}^{+\infty}\frac{(-1)^n}{n + lnn}$$

### $$\sum\limits_{n=1}^{+\infty}\frac{(-1)^n}{n + lnn}$$

### $$\lim_{n\to +\infty} | a_{n}| = \lim_{n\to +\infty} \frac {1}{n + lnn} = 0$$
### $$\sum\limits_{n=1}^{+\infty}| a_{n}| = \frac{1}{1 + ln1}  + \frac{1}{2 + ln2} + \frac{1}{2 + ln2} + ... +  \frac{1}{n + lnn} \\- убывает\,монотонно$$
Знакопостоянный ряд расходится по второму признаку сходимости, следовательно знакопостоянный ряд сходится условно 

__4.__ Исследовать ряд на сходимость, используя признак Раабе:


### $$\sum\limits_{n=1}^{+\infty}\frac{3^n}{2^n}$$

### $$\lim_{n\to +\infty} n\Bigl(\frac{a_n}{a_{n+1}} -1 \Bigr) = \lim_{n\to +\infty} n\Bigl(\frac{3^n}{2^n} \cdot \frac{2^{n+1}}{3^{n+1}} -1 \Bigr) = \lim_{n\to +\infty} n\Bigl(\frac{3^n}{2^n} \cdot \frac{2^n\cdot 2}{3^n \cdot 3} -1 \Bigr) = \lim_{n\to +\infty} n\Bigl(\frac{2}{3} -1 \Bigr) = \lim_{n\to +\infty} n\Bigl(-\frac{1}{3}\Bigr) = -\infty$$

### $$q < 1 - ряд\,расходится $$
