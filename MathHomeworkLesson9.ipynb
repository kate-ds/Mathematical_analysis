__3.__ Численно найти хотя бы один действительный корень системы нелинейных уравнений:
### $$\begin{cases}
   x^2-y^2+3xy^3-2x^2y^2+2x-3y-5=0 \\
   3y^3-2x^2+2x^3y-5x^2y^2+5=0
 \end{cases}$$

```python
from scipy.optimize import fsolve
import math

def equations(p):
    x, y = p
    # Запись системы уравнения вида x+y=2, x-y=0
    return (x**2 - y**2 + 3*x*y**3 - 2*(x**2)*(y**2) + 2*x - 3*y - 5,
            3*y**3 - 2*x**2 + 2*y*x**3 - 5*(x**2)*(y**2) + 5)

# Численное решение системы уравнений
x, y =  fsolve(equations, (10, 10))
print (x, y)
```

__4*.__ Численно найти все $5$ действительных корней.

```python
import numpy as np
from scipy.optimize import fsolve
import math

def equations(p):
    x, y = p
    # Запись системы уравнения вида x+y=2, x-y=0
    return (x**2 - y**2 + 3*x*y**3 - 2*(x**2)*(y**2) + 2*x - 3*y - 5,
            3*y**3 - 2*x**2 + 2*y*x**3 - 5*(x**2)*(y**2) + 5)

x_arr = set()
for x in np.arange (-50, 50, 2):
    for y in np.arange (-50, 50, 2):
        (x, y), info, ier, mesg =  fsolve(equations, (x, y), full_output=True)
        if ier == 1:
            x_arr.add((int(x*10**8)/10**8,int(y*10**8)/10**8))

for ind, el in enumerate(x_arr, 1):
    print (f"solution_{ind}: {el}")
```
