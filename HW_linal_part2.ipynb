__4.__ LU - разложение матрицы

```python
A = np.array([ [1, 2, 3], [2, 16, 21], [4, 28, 73]])
P, L, U = scipy.linalg.lu(A)

print(P)
print(L)
print(U)

print(np.dot(P.transpose(), A) - np.dot(L, U))
```

__5.__ Найти одно из псевдорешений

```python
def Q(x, y, z):
    return (x**2 + y**2 + z**2)

x = np.linspace(-5, 6, 100)
plt.plot(x, Q(x, 10*x - 14, 21*x - 1))
plt.xlabel('x')
plt.ylabel('Q(x)')
plt.show()
```
