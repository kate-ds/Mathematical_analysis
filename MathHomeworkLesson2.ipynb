#### 1. Как соотносятся понятия “множество” и “последовательность”? (в ответе использовать слова типа: часть, целое, общее, частное, родитель, дочерний субъект и т.д.)

Множество - это совокупность каких-либо уникальных элементов. Элементы могут принимать любые значения, могут быть как упорядочены, так и неупорядочены. Последовательность является дочерним элементом множеством.  Последовательность - это упорядоченный пронумерованный набор элементов

#### 2. Прочитать высказывания математической логики, построить их отрицания и установить истинность.

### $$1)\,\,\, \forall y\in[0; 1]:\,\,sgn(y)=1;$$
### $$2)\,\,\, \forall n\in\mathbb{N} > 2:\,\, \exists x, y, z\in\mathbb{N}: x^n = y^n + x^n;$$
### $$3)\,\,\, \forall x\in\mathbb{R} \exists X\in\mathbb{R}: X>x;$$ 
### $$4)\,\,\, \forall x\in\mathbb{C} \nexists y\in\mathbb{C}: x>y||x<y;$$
### $$5)\,\,\, \forall y\in\Bigl[0; \frac{\pi}{2}\Bigr]\,\,\,\exists \varepsilon>0:\,\,\, \sin y<\sin(y+\varepsilon);$$
### $$6)\,\,\, \forall y\in\Bigl[0; \pi\Bigr)\,\,\,\exists \varepsilon>0:\,\,\, \cos y>\cos(y+\varepsilon);$$
### $$7)\,\,\, \exists x: x\in \{\mathbb{N, Z, Q, R, C} \}.$$

1. Выражение ложно, отрицание: $\,\,\, \exists y\in[0; 1]:\,\,sgn(y)\neq1;$
2. Выражение ложно, отрицание: $\,\,\, \exists n\in\mathbb{N} > 2:\,\, \forall x, y, z\in\mathbb{N}: x^n\neq y^n + x^n;$   
(по правилам отрицания вроде так, но по теореме для любых n не найдется решений, я бы поставила знак $\forall $ вначале) 
3. Выражение истинно, отрицание: $\,\,\, \exists X\in\mathbb{R} \forall X\in\mathbb{R}: X\leqslant x$.
4. Выражение ложно, отрицание: $\exists x\in\mathbb{C} \exists y\in\mathbb{C}: x\leqslant y||x\geqslant y;$
5. Выражение истинно, отрицание: $\exists y\in\Bigl[0; \frac{\pi}{2}\Bigr]\,\,\,\forall \varepsilon>0:\,\,\, \sin y\geqslant \sin(y+\varepsilon);$
6. Выражение ложно, отрицание: $\exists y\in\Bigl[0; \pi\Bigr)\,\,\,\forall \varepsilon>0:\cos y\leqslant \cos(y+\varepsilon);$
7. Выражение истинно. Отрицание: $\forall x: x\notin \{\mathbb{N, Z, Q, R, C} \}.$

#### 1. Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.

a = {1, 2, 3, 4}  
b = {2, 3, 4, 7, 8}  
c = {4, 5, 6}  

1. Объединение  

a ∪ b = {1, 2, 3, 4, 7, 8}  
a ∪ c = {1, 2, 3, 4, 5, 6}  
b ∪ c = {2, 3, 4, 5, 6, 7, 8}  

2. Пересечение  

a ⋂ b = {2, 3, 4}  
a ⋂ c = {4}  
b ⋂ c = {4}  

3. Разность  

a \ b = {1}  
b \ a = {7, 8}  
b \ c = {2, 3, 7, 8}  
c \ b = {5, 6}  
a \ c = {1, 2, 3}  
c \ a = {5, 6}  

4. Симметрическая разность  

a Δ b = {1, 7, 8}     
b Δ a = {7, 8, 1}    
b Δ c = {2, 3, 7, 8, 5, 6}    
c Δ b = {5, 6, 2, 3, 7, 8}    
a Δ c = {1, 2, 3, 5, 6}    
c Δ a = {5, 6, 1, 2, 3}   

#### 2. *Выполнить задание 1 на языке Python

```python
a = {1, 4, 5, 6, 'red', 'blue', 'white'}
b = {2, 3, 5, 7, 1, 'black', 'red'}
c = {1, 3, 6, 'orange', 'blue'}
```

```python
# объединение
print('a | b:', a | b, 
      'a | c:', a | c, 
      'b | c:', b | c, sep='\n')
```

```python
# пересечение
print('a & b:', a & b, 
      'a & c:', a & c, 
      'b & c:', b & c, sep='\n')
```

```python
# разность
print('a - b: ', a - b, 
      'a - c: ', a - c, 
      'b - a: ', b - a, 
      'b - c: ', b - c, 
      'c - a: ', c - a, 
      'c - b: ', c - b, sep='\n')
```

```python
# симметрическая разность
print('a ^ b: ', a ^ b, 
      'a ^ c: ', a ^ c, 
      'b ^ a: ', b ^ a, 
      'b ^ c: ', b ^ c, 
      'c ^ a: ', c ^ a, 
      'c ^ b: ', c ^ b, sep='\n')
```
